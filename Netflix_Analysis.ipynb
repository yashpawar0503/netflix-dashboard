## how many movies were added each year?

```python
df.head()
```

```python
df['date_added']=df['date_added'].str.strip()
```

```python
df['date_added']=pd.to_datetime(df['date_added'],format='%B %d, %Y')
```

```python
df['year_added']=df['date_added'].dt.year
```

```python
df.year_added.value_counts()
```

```python
yearadded_names=df.year_added.value_counts().index
```

```python
yearadded_names.astype(str)
```

```python
yearadded_count=df.year_added.value_counts().values
```

```python
yearadded_count
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.lineplot(x=yearadded_names,y=yearadded_count,linewidth=4,color='red')
plt.title('Netflix Additions Over the Years')
plt.style.use('dark_background')
plt.xticks(yearadded_names)
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Observations
1. Rise of netflix happened only after 2015
2. most of the movies or tv shows were added in 2019
3. addition of movies declined after 2019 maybe due to global pandemic

## Countries with most movies or TV Shows

```python
country_names=df[df['country']!='Unknown']['country'].value_counts().head(10).index
```

```python
country_values=df[df['country']!='Unknown']['country'].value_counts().head(10).values
```

```python
plt.style.use('dark_background')

sns.barplot(x=country_names,y=country_values,color='red')
plt.title('Countries with the no of Netflix Shows')
plt.ylabel('Count of shows/Movies')
plt.xticks(rotation=45)
plt.show()
```

## Observations
1. United states has the highest no of Tv Shows or Movies on Netflix
2. India ranks 2nd followed by UK at 3rd


## Top Genres

```python
df.listed_in.value_counts()
```

```python
genre_series = df['listed_in'].str.split(',').explode().str.strip()
```

```python
genre_counts=genre_series.value_counts().values
```

```python
genre_names=genre_series.value_counts().index
```

```python
plt.rcParams['figure.figsize']=(20,15)
plt.style.use('dark_background')
sns.barplot(x=genre_names,y=genre_counts,color='red')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('number of titles')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
